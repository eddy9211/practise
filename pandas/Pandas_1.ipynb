Above dataset  has an in-built index, which pandas did not pick up on automatically. To make pandas use that column for the index (instead of creating a new one from scratch), we may specify and use an index_col.

```python
w_r = pd.read_csv('winemag-data-130k-v2.csv',index_col = 0)
w_r.head()
```
